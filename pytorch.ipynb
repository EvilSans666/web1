# K-Means

```python
import matplotlib.pyplot as plt
```

```python
N = 100
x1 = torch.randn(N, 2)
x2 = torch.randn(N, 2) + torch.Tensor([3, 3])
x3 = torch.randn(N, 2) + torch.Tensor([-3, 3])
```

```python
X = torch.cat([x1, x2, x3], dim=0)
print(X.shape)
```

```python
for i in range(3):
    plt.scatter(X[i*N:(i+1)*N, 0], X[i*N:(i+1)*N, 1])
plt.show()
```
